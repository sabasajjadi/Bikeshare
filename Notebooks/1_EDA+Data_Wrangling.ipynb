```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df= pd.read_csv("hourly.csv")
```

```python
print("Our data have {} rows and {} columns" .format(df.shape[0], df.shape[1]))
```

```python
df.describe()
```

```python
fig, ax =plt.subplots(figsize=(15, 15))
plt.title("Correlation Plot")
sns.heatmap(df.corr(), mask=np.zeros_like(df.corr(), dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax, annot=True,linewidths=3)
plt.show()
```

```python
#removing irrelevant fetaures basedon correlation
columns=["instant", "dteday", "yr", "mnth", "hr", "holiday", "weekday", "workingday", "temp"]

df= df.drop(columns, axis=1)
```

```python
print("After removing some irrelevant features our data have {} rows and {} columns" .format(df.shape[0], df.shape[1]))
```

```python
fig, ax =plt.subplots(figsize=(15, 15))
plt.title("Correlation Plot")
sns.heatmap(df.corr(), mask=np.zeros_like(df.corr(), dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax, annot=True,linewidths=3)
plt.show()
```
